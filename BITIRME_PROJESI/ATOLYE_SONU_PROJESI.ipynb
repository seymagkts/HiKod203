### Sürdürülebilirlik (Sustainability)

CARBON EMISSION DATASET FEATURES:

* 'Body Type': Vücut tipi (Body type)

* 'Sex': Cinsiyet (Gender)

* 'Diet': Diyet (Diet)

* 'How Often Shower': Duş alma sıklığı (Frequency of showering)

* 'Heating Energy Source': Konut ısıtma enerjisi (Residential heating energy)
    
* 'Transport': Ulaşım tercihi (Transportation preference)
    
* 'Vehicle Type': Araç yakıt tipi (Vehicle fuel type)

* 'Social Activity': Sosyal etkinliklere katılma sıklığı (Frequency of participating in social activities)

* 'Monthly Grocery Bill': Market alışverişine harcanan aylık tutar, dolar cinsinden (Monthly amount spent on groceries, in dollars)

* 'Frequency of Traveling by Air': Son ayda uçak kullanma sıklığı (Frequency of using aircraft in the last month)

* 'Vehicle Monthly Distance Km':  Aracın son ayda kat ettiği kilometre (The kilometers traveled by vehicle in the last month)

* 'Waste Bag Size': Çöp torbasının boyutu (Size of the garbage bag)

* 'Waste Bag Weekly Count': Son hafta içinde atılan çöp miktarı (The amount of garbage thrown away in the last week)

* 'How Long TV PC Daily Hour': TV veya PC karşısında geçirilen günlük süre (Daily time spent in front of TV or PC)

* 'How Many New Clothes Monthly': Aylık satın alınan kıyafet sayısıdır (Number of clothes purchased monthly)

* 'How Long Internet Daily Hour': Günlük olarak internette geçirilen süre (Time spent on the Internet daily)
    
* 'Energy efficiency': Enerji tasarruflu cihazlar satın almayı önemseyip önemsemediğiniz (Whether or not you care about purchasing energy efficient devices)

* 'Recycling': Geri dönüştürdüğü atıklardır (The wastes it recycles)

* 'Cooking_With': Yemek pişirmede kullanılan cihazlar (Devices used in cooking)

* 'CarbonEmission': Bağımlı değişken, toplam karbon emisyonları, kg (Dependent variable, total carbon emissions)

KAYNAK (SOURCE): https://www.kaggle.com/datasets/dumanmesut/individual-carbon-footprint-calculation

## MODEL

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,r2_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
binary_cols = [col for col in df_copy.columns if df_copy[col].dtypes == "O" and df_copy[col].nunique() == 2]

def label_encoder(dataframe, binary_col):
    labelencoder = LabelEncoder()
    dataframe[binary_col] = labelencoder.fit_transform(dataframe[binary_col])
    return dataframe

for col in binary_cols:
    df_copy = label_encoder(df_copy, col)
    
# kategorik verileri sayısal değerlere dönüştürme yani label encoding işlemi
```

```python
cat_cols = [col for col in df_copy if col not in binary_cols and col not in ["Carbon_Emission"]]

def one_hot_encoder(dataframe, categorical_cols, drop_first=True):
    dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=drop_first)
    return dataframe

ohe_cols = [col for col in df_copy.columns if 10 >= df_copy[col].nunique() > 2]

df_copy = one_hot_encoder(df_copy, ohe_cols, True)

# kategorik verileri ikili (binary) vektörlere dönüştürme yani one hot encoding işlemi
```

```python
X = df_copy.drop(["Carbon_Emission","Recycling","Cooking_With"], axis=1)
y = df_copy["Carbon_Emission"]
# bağımlı ve bağımsız değişkenleri ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 17)
# veri setini test ve train seti olarak bölme
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test =ss.transform(X_test)

# ortalama değerleri 0 ve standart sapmaları 1 olacak şekilde dönüştürme yani standartlaştırma işlemi
```

```python
def plot_importance(model, features, num=len(X)):
    feature_imp = pd.DataFrame({'Değerler': model.feature_importances_, 'Özellikler': features.columns})
    plt.figure(figsize=(10, 8))
    sns.set(font_scale=1)
    sns.barplot(x="Değerler", y="Özellikler", data=feature_imp.sort_values(by="Değerler",
                                                                     ascending=False)[0:num])
    plt.title('Model bazında özelliklerin önem sırası')
    plt.tight_layout()
    plt.show()
    
# kurulan modele göre en önemli özellikleri sıralayan ve görselleştiren fonksiyon
```

```python
models = [('KNN', KNeighborsRegressor()),
          ('CART', DecisionTreeRegressor()),
          ('RF', RandomForestRegressor()),
          ('GBM', GradientBoostingRegressor()),
          ("XGBoost", XGBRegressor()),
          ("LightGBM", LGBMRegressor(verbose=-1)),
          ("CatBoost", CatBoostRegressor(verbose=False))]
# model basarılarını kıyaslamak için birden çok modeli tanımlama
```

```python
for name, regressor in models:
    rmse = np.mean(np.sqrt(-cross_val_score(regressor, X, y, cv=5, scoring="neg_mean_squared_error")))
    print(f"RMSE: {round(rmse, 4)} ({name}) ")
    
# modellerin rmse değerleri
# cross_val_score fonksiyonunu modelin ne kadar iyi çalıştığını RMSE değeri ve 5 katlı doğrulama ile 
# tespit etmek amacıyla kullandım
```

***

### Regresyon Başarı Metrikleri
* R2 Başarı Metriği
###### R2, bağımsız değişkenlerin bağımlı değişkeni ne kadar iyi açıkladığını gösteren ölçüdür.
###### 0 ile 1 arasında değer alır. 1'e yaklaşan değerler, bağımsız değişkenlerin bağımlı değişkeni iyi açıkladığını gösterir.

* RMSE Başarı Metriği
###### RMSE, gerçek değerler ile modelin tahmin ettiği değerler arasındaki farkların karelerinin ortalamasının kareköküdür.
###### Daha düşük RMSE değeri, modelin tahminlerinin gerçek değerlere daha yakın olduğunu gösterir.

### En iyi RMSE değerini veren iki model
#### RMSE: 186.4578 (LightGBM)
#### RMSE: 152.9947 (CatBoost)

### LIGHT GBM HIPERPARAMETRE OPTIMIZASYONU

Modeli daha iyi hale getirmek için doğru hiperparametreleri seçme sürecidir ve genellikle deneme-yanılma (trial-and-error) yöntemlerini içerir.

```python
lightgbm_params =  {"learning_rate":[0.01,0.1],
                 'n_estimators': [100,350,500,1000],
                 "colsample_bytree":[1,0.5,0.7]}

# light gbm modeli üzerinde parametre optimizasyonunda deneyeceğim parametreleri tanımladım
```

```python
lightgbm_best_grid = GridSearchCV(lightgbm_model,
                             lightgbm_params,
                             cv=5,
                              n_jobs=-1,
                             verbose=1).fit(X_train,y_train)

# grid search cv ile optimizasyon işlemini 5 katlı çapraz doğrulama olacak şekilde başlattım
```

```python
lightgbm_best_grid.best_params_
# verilen parametreler arasında en iyi sonucu veren parametreler
```

```python
# {'colsample_bytree': 0.7, 'learning_rate': 0.1, 'n_estimators': 500}
```

```python
lgbm_final = LGBMRegressor(**lightgbm_best_grid.best_params_,verbose=-1).fit(X_train,y_train)
# bu parametrelere göre yeni bir final modeli kurdum
```

### TRAIN RMSE, R2

```python
y_pred_lgbm_opt = lgbm_final.predict(X_train)
np.sqrt(mean_squared_error(y_train, y_pred_lgbm_opt)) 
# final modelinin train setindeki rmse değeri
```

```python
lgbm_final.score(X_train, y_train)  
# final modelinin train setindeki r2 değeri
```

### TEST RMSE, R2

```python
y_pred_lgbm_opt = lgbm_final.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred_lgbm_opt)) 
# final modelinin test setindeki rmse değeri
```

```python
lgbm_test_opt_r2 = lgbm_final.score(X_test, y_test)  
lgbm_test_opt_r2 
# final modelinin test setindeki r2 değeri
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_lgbm, color='#3b9fff', label='Light GBM Tahminleri', marker='o',s=100,edgecolors="#9bceff")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='--', linewidth=2, label='Doğru')
plt.title(f'Light GBM: Optimizasyonsuz {round(lgbm_test_r2,4)}', fontsize=15)
plt.xlabel('Gerçek Değerler', fontsize=12)
plt.ylabel('Tahmin Edilen Değerler', fontsize=12)
plt.legend(fontsize=12)

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_lgbm_opt, color='#24ff2d', label='Light GBM Tahminleri', marker='o', s=100,edgecolors="#c9ffcb")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='--', linewidth=2, label='Doğru')
plt.title(f'Light GBM: Optimizasyonlu {round(lgbm_test_opt_r2,4)}', fontsize=15)
plt.xlabel('Gerçek Değerler', fontsize=12)
plt.ylabel('Tahmin Edilen Değerler', fontsize=12)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

# light gbmin ilk ve son halindeki başarının artışını görselleştirdim, scatter ile.
```
