### Sürdürülebilirlik (Sustainability)

CARBON EMISSION DATASET FEATURES:

* 'Body Type': Vücut tipi (Body type)

* 'Sex': Cinsiyet (Gender)

* 'Diet': Diyet (Diet)

* 'How Often Shower': Duş alma sıklığı (Frequency of showering)

* 'Heating Energy Source': Konut ısıtma enerjisi (Residential heating energy)
    
* 'Transport': Ulaşım tercihi (Transportation preference)
    
* 'Vehicle Type': Araç yakıt tipi (Vehicle fuel type)

* 'Social Activity': Sosyal etkinliklere katılma sıklığı (Frequency of participating in social activities)

* 'Monthly Grocery Bill': Market alışverişine harcanan aylık tutar, dolar cinsinden (Monthly amount spent on groceries, in dollars)

* 'Frequency of Traveling by Air': Son ayda uçak kullanma sıklığı (Frequency of using aircraft in the last month)

* 'Vehicle Monthly Distance Km':  Aracın son ayda kat ettiği kilometre (The kilometers traveled by vehicle in the last month)

* 'Waste Bag Size': Çöp torbasının boyutu (Size of the garbage bag)

* 'Waste Bag Weekly Count': Son hafta içinde atılan çöp miktarı (The amount of garbage thrown away in the last week)

* 'How Long TV PC Daily Hour': TV veya PC karşısında geçirilen günlük süre (Daily time spent in front of TV or PC)

* 'How Many New Clothes Monthly': Aylık satın alınan kıyafet sayısıdır (Number of clothes purchased monthly)

* 'How Long Internet Daily Hour': Günlük olarak internette geçirilen süre (Time spent on the Internet daily)
    
* 'Energy efficiency': Enerji tasarruflu cihazlar satın almayı önemseyip önemsemediğiniz (Whether or not you care about purchasing energy efficient devices)

* 'Recycling': Geri dönüştürdüğü atıklardır (The wastes it recycles)

* 'Cooking_With': Yemek pişirmede kullanılan cihazlar (Devices used in cooking)

* 'CarbonEmission':Bağımlı değişken, toplam karbon emisyonları, kg (Dependent variable, total carbon emissions)

KAYNAK (SOURCE): https://www.kaggle.com/datasets/dumanmesut/individual-carbon-footprint-calculation

## MODEL

```python
from sklearn.preprocessing import LabelEncoder,RobustScaler,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, cross_validate, RandomizedSearchCV, validation_curve,cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error,r2_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
binary_cols = [col for col in df_copy.columns if df_copy[col].dtypes == "O" and df_copy[col].nunique() == 2]

def label_encoder(dataframe, binary_col):
    labelencoder = LabelEncoder()
    dataframe[binary_col] = labelencoder.fit_transform(dataframe[binary_col])
    return dataframe

for col in binary_cols:
    df_copy = label_encoder(df_copy, col)
```

```python
cat_cols = [col for col in df_copy if col not in binary_cols and col not in ["Carbon_Emission"]]

def one_hot_encoder(dataframe, categorical_cols, drop_first=True):
    dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=drop_first)
    return dataframe

ohe_cols = [col for col in df_copy.columns if 10 >= df_copy[col].nunique() > 2]

df_copy = one_hot_encoder(df_copy, ohe_cols, True)
```

```python
num_cols = df_copy.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    if col != "Carbon_Emission":
        transformer = StandardScaler().fit(df_copy[[col]])
        df_copy[col] = transformer.transform(df_copy[[col]])
```

```python
X = df_copy.drop(["Carbon_Emission","Recycling","Cooking_With"], axis=1)
y = df_copy["Carbon_Emission"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
```

```python
def plot_importance(model, features, num=len(X)):
    feature_imp = pd.DataFrame({'Değerler': model.feature_importances_, 'Özellikler': features.columns})
    plt.figure(figsize=(10, 8))
    sns.set(font_scale=1)
    sns.barplot(x="Değerler", y="Özellikler", data=feature_imp.sort_values(by="Değerler",
                                                                     ascending=False)[0:num])
    plt.title('Model bazında özelliklerin önem sırası')
    plt.tight_layout()
    plt.show()
    
# kurulan modele göre en önemli özellikleri sıralayan ve görselleştiren fonksiyon
```

```python
models = [#('LR', LogisticRegression()),
          ('KNN', KNeighborsRegressor()),
          ('CART', DecisionTreeRegressor()),
          ('RF', RandomForestRegressor()),
          ('GBM', GradientBoostingRegressor()),
          ("XGBoost", XGBRegressor()),
          ("LightGBM", LGBMRegressor()),
          ("CatBoost", CatBoostRegressor(verbose=False))]
```

```python
for name, regressor in models:
    rmse = np.mean(np.sqrt(-cross_val_score(regressor, X, y, cv=5, scoring="neg_mean_squared_error")))
    print(f"RMSE: {round(rmse, 4)} ({name}) ")
```

### CATBOOST MODEL

```python
catboost_model = CatBoostRegressor().fit(X_train, y_train)
```

### TRAIN RMSE, R2

```python
y_pred_catb = catboost_model.predict(X_train)
np.sqrt(mean_squared_error(y_train, y_pred_catb)) 
```

```python
catboost_model.score(X_train, y_train) 
```

### TEST RMSE, R2

```python
y_pred_catb = catboost_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred_catb))
```

```python
catboost_model.score(X_test, y_test)
```

```python
plot_importance(catboost_model, X)
```
